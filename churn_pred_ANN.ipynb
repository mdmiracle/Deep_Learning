```python
def print_unique_value(df):
    for col in df.columns:
        if df[col].dtype=='object':
            print(f'{col}: ,{df[col].unique()}')
            #print(col)
print_unique_value(df)
```

```python
df.replace('No internet service','No',inplace=True)
df.replace('No phone service','No',inplace=True)
```

```python
print_unique_value(df)
```

```python
yes_no_col = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in yes_no_col:
    df[col].replace({'Yes':1,'No':0},inplace=True) 

for col in df.columns:
       print(f'{col}: ,{df[col].unique()}')
```

```python
df.gender.replace({'Male':1,'Female':0},inplace=True)
df.gender.unique()
```

```python
#pd.get_dummies(df.InternetService)
df = pd.get_dummies(data=df,columns=['InternetService','Contract','PaymentMethod'])
```

```python
df.sample(5)
```

```python
for col in df.columns:
       print(f'{col}: ,{df[col].unique()}')
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['tenure','MonthlyCharges','TotalCharges']] = scaler.fit_transform(df[['tenure','MonthlyCharges','TotalCharges']])
```

```python
df.sample(5)
```

```python
x = df.drop(['Churn'],axis = 1)
y = df['Churn']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
x_train.shape,y_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
y_pred[:5]
```

```python
y_test
```

```python
y_pred_mod = []
for value in y_pred:
    if value>0.5:
        y_pred_mod.append(1)
    else:
        y_pred_mod.append(0)
```

```python
y_pred_mod[:5]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(classification_report(y_test,y_pred_mod))
```

```python
confusion_matrix(y_test,y_pred_mod)
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred_mod)

plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,fmt='d')
plt.xlable('predictions')
plt.ylable('true')
```
