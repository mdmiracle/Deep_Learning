```python
import tensorflow as tf
import keras
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
from keras.preprocessing.image import ImageDataGenerator

from keras.utils import img_to_array,array_to_img,load_img
```

```python
img = load_img((r"C:\Users\91779\Downloads\golden-retriever-royalty-free-image-506756303-1560962726.jpg"))
x = img_to_array(img)
img
```

```python
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)
```

```python
x = x.reshape(-1,605,1200,3)
x.shape
```

```python
i=0
for batch in datagen.flow(x,batch_size=1,save_to_dir='imajes_cat',save_prefix='cat',save_format='jpeg'):
    i +=1
    if i>20:
        break

```
